```python
import cv2
import matplotlib.pyplot as plt
```

```python
img_pepsi = cv2.imread('C:\Programming\Python\OpenCV\pepsi_img.PNG')

img_pepsi = cv2.cvtColor(img_pepsi, cv2.COLOR_BGR2RGB)

plt.imshow(img_pepsi)

```

When you read an image using cv2.imread(), OpenCV loads the image in BGR format by default.

When you write an image using cv2.imwrite(), OpenCV expects the image data in BGR format.

Hence the need for line #2 above

```python
b, g, r = cv2.split(img_pepsi)

plt.figure(figsize = [20, 4]) # width ---> 20 ; height ---> 4        Any value greater than mentioned height and width value will not result in any change in the output. But any lessser, will.

plt.subplot(151)
plt.imshow(img_pepsi)
plt.title('Original image')

plt.subplot(152)
plt.imshow(r, cmap = 'gray')
plt.title('Red Channel')

plt.subplot(153)
plt.imshow(g, cmap = 'gray')
plt.title('Green Channel')

plt.subplot(154)
plt.imshow(b, cmap = 'gray')
plt.title('Blue Channel')


merged = cv2.merge((b, g, r))
plt.subplot(155)
plt.imshow(merged)
plt.title('Merged Output')


```
