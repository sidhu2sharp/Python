```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('C:/Programming/Python/OpenCV/WIN_20230728_16_19_21_Pro.jpg', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
cv2.imwrite("college.jpg", img)

```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray, cmap = 'gray')
```

```python
r, g, b = cv2.split(img)
img_new_gray = 0.33 * r + 0.33 * g + 0.33 * b
plt.imshow(img_new_gray, cmap = 'gray')
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
plt.imshow(img_hsv)
```

```python
img_luv = cv2.cvtColor(img, cv2.COLOR_RGB2LUV)
plt.imshow(img_luv)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
img_inv = 255 - img_gray
plt.imshow(img_inv, cmap = 'gray')
```
